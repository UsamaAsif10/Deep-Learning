```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_squared_error, mean_absolute_percentage_error, r2_score
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import SimpleRNN, LSTM, GRU, Dense, Dropout
from tensorflow.keras.optimizers import Adam

# Load and preprocess the dataset
data = pd.read_csv('Downloads/archive/Google_Stock_Price_Train.csv')
for col in ['Open', 'High', 'Low', 'Close', 'Volume']:
    data[col] = data[col].replace({',': ''}, regex=True).astype(float)
features = ['Open', 'High', 'Low', 'Close', 'Volume']
dataset = data[features].values

# Scale the data
scaler = MinMaxScaler(feature_range=(0, 1))
scaled_data = scaler.fit_transform(dataset)

# Create sequences using 30 days as the look-back window
look_back = 30
X, y = create_sequences(scaled_data, look_back)

# Function to create sequences with all features
def create_sequences(data, look_back, target_col=0):
    X, y = [], []
    for i in range(look_back, len(data)):
        X.append(data[i-look_back:i, :])  # Use all features for input
        y.append(data[i, target_col])  # Predict the target column (default: 'Open')
    return np.array(X), np.array(y)

# Split data into training, validation, and test sets
train_size = int(0.8 * len(X))
X_train_full, y_train_full = X[:train_size], y[:train_size]
X_test, y_test = X[train_size:], y[train_size:]

val_size = int(0.2 * len(X_train_full))
X_train, y_train = X_train_full[:-val_size], y_train_full[:-val_size]
X_val, y_val = X_train_full[-val_size:], y_train_full[-val_size:]

# Function to build models
def build_model(model_type, input_shape, learning_rate=0.001, dropout_rate=0.2):
    model = Sequential()
    if model_type == 'RNN':
        model.add(SimpleRNN(50, activation='tanh', return_sequences=False, input_shape=input_shape))
    elif model_type == 'LSTM':
        model.add(LSTM(50, return_sequences=False, input_shape=input_shape))
    elif model_type == 'GRU':
        model.add(GRU(50, return_sequences=False, input_shape=input_shape))
    model.add(Dropout(dropout_rate))
    model.add(Dense(25, activation='relu'))
    model.add(Dense(1))  # Output layer
    model.compile(optimizer=Adam(learning_rate=learning_rate), loss='mean_squared_error')
    return model

from sklearn.metrics import mean_squared_error

# Hyperparameter grid
hyperparams = {
    'learning_rate': [0.001, 0.01], 
    'dropout_rate': [0.2, 0.3], 
    'batch_size': [16, 32], 
    'epochs': [10, 20]
}

# Initialize results storage
results = {}
best_models = {}

# Hyperparameter tuning for each model
for model_type in ['RNN', 'LSTM', 'GRU']:
    print(f"\nTuning hyperparameters for {model_type}...")

    best_val_rmse = float('inf')  # Start with an infinitely large RMSE
    best_model = None
    best_params = None

    # Loop through all combinations of hyperparameters
    for learning_rate in hyperparams['learning_rate']:
        for dropout_rate in hyperparams['dropout_rate']:
            for batch_size in hyperparams['batch_size']:
                for epochs in hyperparams['epochs']:
                    print(f"Training {model_type} with lr={learning_rate}, dropout={dropout_rate}, batch_size={batch_size}, epochs={epochs}")

                    # Build the model with the current hyperparameters
                    model = build_model(model_type, (X_train.shape[1], X_train.shape[2]), 
                                        learning_rate=learning_rate, dropout_rate=dropout_rate)

                    # Train the model
                    model.fit(X_train, y_train, validation_data=(X_val, y_val), epochs=epochs, batch_size=batch_size, verbose=1)

                    # Evaluate on the validation set
                    val_predictions = model.predict(X_val)
                    val_rmse = np.sqrt(mean_squared_error(y_val, val_predictions))  # RMSE on validation set

                    # Check if the current model is better than the best found so far
                    if val_rmse < best_val_rmse:
                        best_val_rmse = val_rmse
                        best_model = model
                        best_params = {
                            'learning_rate': learning_rate,
                            'dropout_rate': dropout_rate,
                            'batch_size': batch_size,
                            'epochs': epochs
                        }

    print(f"Best parameters for {model_type}: {best_params}")
    print(f"Best validation RMSE: {best_val_rmse:.4f}")

    # Store the best model and its parameters for later evaluation
    best_models[model_type] = best_model
    results[model_type] = best_params

# Now we evaluate the best model on the test set
for model_type, model in best_models.items():
    print(f"\nEvaluating the best model for {model_type} on the test set...")
    
    # Predict test set results
    test_predictions = model.predict(X_test)

    # Inverse scaling of the predictions
    test_predictions = scaler.inverse_transform(np.concatenate([test_predictions, np.zeros((test_predictions.shape[0], scaled_data.shape[1]-1))], axis=1))[:, 0]
    y_test_actual = scaler.inverse_transform(np.concatenate([y_test.reshape(-1, 1), np.zeros((y_test.shape[0], scaled_data.shape[1]-1))], axis=1))[:, 0]

    # Calculate metrics for the test set
    test_rmse = np.sqrt(mean_squared_error(y_test_actual, test_predictions))
    test_mape = mean_absolute_percentage_error(y_test_actual, test_predictions) * 100
    test_r2 = r2_score(y_test_actual, test_predictions)
    test_accuracy = 100 - test_mape  # Calculate Test Accuracy

    # Store metrics in the results dictionary
    results[model_type].update({
        'Test RMSE': test_rmse,
        'Test MAPE': test_mape,
        'Test R2': test_r2,
        'Test Accuracy (%)': test_accuracy
    })

    # Print results for this model
    print(f"{model_type} Test Metrics:")
    print(f"Test RMSE: {test_rmse:.4f}")
    print(f"Test MAPE: {test_mape:.4f}")
    print(f"Test R2: {test_r2:.4f}")
    print(f"Test Accuracy (%): {test_accuracy:.4f}")

# Print final results summary
print("\nFinal Model Performance Summary:")
print(f"{'Model':<10} {'Test RMSE':<15} {'Test MAPE':<15} {'Test R2':<10} {'Test Accuracy (%)':<20}")
print("="*60)

    # Plot predictions vs actual values
    plt.figure(figsize=(14, 8))
    plt.plot(range(len(y_train_actual)), y_train_actual, label='Training Actual', color='blue')
    plt.plot(range(len(y_train_actual), len(y_train_actual) + len(y_val_actual)), y_val_actual, label='Validation Actual', color='orange')
    plt.plot(range(len(y_train_actual) + len(y_val_actual), len(y_train_actual) + len(y_val_actual) + len(y_test_actual)), y_test_actual, label='Test Actual', color='red')

    plt.plot(range(len(train_predictions)), train_predictions, label='Training Predictions', linestyle='dashed', color='cyan')
    plt.plot(range(len(y_train_actual), len(y_train_actual) + len(val_predictions)), val_predictions, label='Validation Predictions', linestyle='dashed', color='gold')
    plt.plot(range(len(y_train_actual) + len(y_val_actual), len(y_train_actual) + len(y_val_actual) + len(test_predictions)), test_predictions, label='Test Predictions', linestyle='dashed', color='green')

    plt.title(f'{model_type} - Actual vs Predicted Stock Prices')
    plt.xlabel('Time Steps')
    plt.ylabel('Stock Prices')
    plt.legend()
    plt.grid()
    plt.show()
```
